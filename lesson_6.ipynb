```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"Ice Cream Ratings.csv")
df.set_index('Date', inplace=True)
df
```

```python
# Таким мы можем управлять стилем отобраежния графиков
print(plt.style.available)
plt.style.use('tableau-colorblind10')
```

```python
# нам выведет линейный график на основе индекса Date
df.plot()
```

```python
# Выведет 3 отдельных графика так как есть 3 коллонный которые относятся к индексу, также добавляет загаловок к графику
df.plot(kind='line', subplots=True, title='Ice Cream Rating')
```

```python
# Даем лейблы(названия) для двух осей
df.plot(kind='line', title='Ice Cream Rating', xlabel='Daily Ratings', ylabel='Scores')
```

```python
# Мы тут берем колонну Flavor Rating и создаем бар чарт, если бы было без него то оно для всех в той же
# колоннах с тремя цветами так как указан аргумент stascked, а если без него то на линий по три колонки для каждой даты
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
# Тут тоже самое но горизонтально и не указывается kind так как .barh сам по себе метод bar
# Также можно выделить только одну коллонну как и до этог делали
df.plot.barh(stacked=True)
```

```python
# Тут у нас скаттер плот диаграмма и оно не работает без аргументов x,y названия осьей,
# также мы можем менять размер точки
# И цвет по своему желанию
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500, c='Green')
```

```python
# Гистограммы, от значения bins зависит сколько разделений будет в x осьи и по дефолту стоит 10
df.plot.hist(bins=20)
```

```python
df.plot.box()
```

```python
# Можно везде указывать figsize чтобы график был большего размера
df.plot.area(figsize=(10, 5))
```

```python
# Это всеми знакомы график и также казав только subplots можно увидить статистику для каждой коллоны
# Указав y равной какой-то колонне то мы выводим инфу только про нее
df.plot.pie(subplots=True, y='Flavor Rating', figsize=(10, 6))
```
